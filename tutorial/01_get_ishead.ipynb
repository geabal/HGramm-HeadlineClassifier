# HGramm : get_isHead
get_isHead()는 HGramm으로 문서 내 각 문장이 표제어일 확률을 예측하는 메소드입니다.  
HGramm은 feature로 mecab pos 분포를 사용하므로, 사용 전에 mecab 경로를 지정해주어야 합니다.

```python
from HGramm import HGramm

hg = HGramm()

# mecab 사용을 위한 dicpath 경로 설정
mecab_path = 'C:/mecab/mecab-ko-dic'
hg.set_mecab(mecab_path=mecab_path)

```

```python
#get_isHead()에서 doc_list는 List[Dict]로 주어져야 합니다.
print(type(cursor))
print(type(cursor[0]))
# 본문 컬럼 이름: text
# id 컬럼 이름: url
print(cursor[0])
```

```python
from time import time
start = time()
is_head = hg.get_isHead(cursor, text_col='text', id_col='url')
print(len(is_head))
end = time()
print(f'소요시간 : {end-start} sec')
```

```python
# get_isHead는 기본적으로 문장 단위로 예측합니다.
# is_head는 0~1 사이 float 값을 가지며, 1에 가까울수록 표제어일 확률이 높습니다. 반대로 0에 가까울수록 본문 문장일 가능성이 높습니다.
sent_df = hg.sent_df
print(f"sentence: {sent_df.loc[0,'sent']}\nis_head score: {sent_df.loc[0,'is_head']}")
```

```python
# 문장단위로 예측된 is_head 결과를 문서 단위로 묶어 보려면 sent2doc()을 사용하면 됩니다.
doc_df = hg.sent2doc()
doc_df
```
